# Demo subplot

It is often necessairy to present data (images in this example) in a subplot of multiple graphs. This effect can (and sometimes should) be achieved with LaTex instead, but here is an example of 3x2 matplotlib subplot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

we start by creating example images

```python
image_list = [] # we create a empty image list

# adding original lena image to the list
marvin = plt.imread('data/marvin_clean.png')
image_list.append(marvin)

# we create our other images based on zooms on lena
cutsize = 50 # we remove cutsize pixels per itterations on each sides

for i in range(5):
    marvin = image_list[i]
    marvin = marvin[cutsize:-cutsize, cutsize:-cutsize]
    image_list.append(marvin)
```

Now let's create our figure of 6 images.

Finally, we export this image as whatever format we want. In general .svg, .pdf or .png are the only sensible choices. Because images (instead of plots) can be very large .svg and .pdf files, we use .png here.

```python
plt.figure(figsize=[8, 6])

for i in range(len(image_list)):
    im = image_list[i]
    plt.subplot(2, 3, i+1)
    plt.imshow(im)
    plt.axis('off')
    plt.title('Zoom step {}'.format(i))
    
plt.tight_layout() # this is optional, but I tend to use it very often
# plt.savefig('demo_subplot.png', dpi = 196) # dpi can be ajusted to increase .png resolution
```
